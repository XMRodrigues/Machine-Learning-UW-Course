```python
import turicreate
from turicreate import SArrayBuilder
```

```python
products = turicreate.SFrame('amazon_baby.sframe')
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
products['count_words'] = turicreate.text_analytics.count_words(products['review'])
```

```python
products.head()
```

```python
def awesome_count(word):
    if 'awesome' in word:
        return word['awesome']
    else:
        return 0
    
def great_count(word):
    if 'great' in word:
        return word['great']
    else:
        return 0

def fantastic_count(word):
    if 'fantastic' in word:
        return word['fantastic']
    else:
        return 0

def amazing_count(word):
    if 'amazing' in word:
        return word['amazing']
    else:
        return 0

def love_count(word):
    if 'love' in word:
        return word['love']
    else:
        return 0

def horrible_count(word):
    if 'horrible' in word:
        return word['horrible']
    else:
        return 0

def bad_count(word):
    if 'bad' in word:
        return word['bad']
    else:
        return 0

def terrible_count(word):
    if 'terrible' in word:
        return word['terrible']
    else:
        return 0

def awful_count(word):
    if 'awful' in word:
        return word['awful']
    else:
        return 0

def wow_count(word):
    if 'wow' in word:
        return word['wow']
    else:
        return 0

def hate_count(word):
    if 'hate' in word:
        return word['hate']
    else:
        return 0
```

```python
function = selected_words[0]+'_count'
eval(function(products['count_words'][0]))
```

```python
products['awesome'] = products['count_words'].apply(awesome_count)
products['great'] = products['count_words'].apply(great_count)
products['fantastic'] = products['count_words'].apply(fantastic_count)
products['amazing'] = products['count_words'].apply(amazing_count)
products['love'] = products['count_words'].apply(love_count)
products['horrible'] = products['count_words'].apply(horrible_count)
products['bad'] = products['count_words'].apply(bad_count)
products['terrible'] = products['count_words'].apply(terrible_count)
products['awful'] = products['count_words'].apply(awful_count)
products['wow'] = products['count_words'].apply(wow_count)
products['hate'] = products['count_words'].apply(hate_count)
```
